```python
cross_tab = pd.crosstab(Terror.Year, Terror.Region)
regions = cross_tab.columns
colormap = plt.cm.get_cmap('tab20', len(regions))  
area_plot = cross_tab.plot(kind='area', figsize=(15, 6), colormap=colormap)
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.legend(title='Region', labels=regions)
plt.show()
```

```python
Terror['Wounded'] = Terror['Wounded'].fillna(0).astype(int)
Terror['Killed'] = Terror['Killed'].fillna(0).astype(int)
Terror['casualties'] = Terror['Killed'] + Terror['Wounded']
```

```python
Terrordata1 = Terror.sort_values(by='casualties',ascending=False)[:50]
```

```python
heat=Terrordata1.pivot_table(index='Country',columns='Year',values='casualties')
heat.fillna(0,inplace=True)
plt.figure(figsize=(12, 8))
sns.heatmap(heat, cmap='YlOrRd', annot=True, fmt='g')
plt.title('Casualties by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
colorscale = [[0, '#f7fbff'],[.5, '#3182bd'],[.5, '#fee090'],[1, '#e34a33']]  
heatmap = go.Heatmap(
    z=heat.values,            
    x=heat.columns,         
    y=heat.index,          
    colorscale=colorscale    
)
data = [heatmap]
layout = go.Layout(
    title='Top 40 Worst Terror Attacks in History from 1982 to 2016',
    xaxis=dict(ticks='', nticks=20),
    yaxis=dict(ticks='')
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='heatmap', show_link=False)
```

```python
killData = Terror.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))
```

```python
attackData = Terror.loc[:,'AttackType']
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillData.head()
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python

labels = typeKillFormatData.columns.tolist() 
transpoze = typeKillFormatData.T 

fig, ax = plt.subplots(figsize=(20, 20))
plt.title('Types of terrorist attacks that cause deaths', fontsize=20)

for idx, (label, column) in enumerate(transpoze.iteritems()):
    wedges, texts, autotexts = ax.pie(column, autopct='%.2f%%', startangle=90)
    ax.axis('equal')
    ax.legend(wedges, labels, loc='upper right', bbox_to_anchor=(1.3, 0.9), fontsize=15)
    ax.set_title(f'{label} - {column.sum()}')

plt.show()
```

**The causes of death in these attacks can be attributed to armed assaults and bombings/explosions, which account for a significant 77% of the total fatalities. This alarming statistic underscores why these methods are frequently employed in terrorist actions. It serves as a poignant reminder of the immense danger posed by weapons and explosives to global security.
